```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
datos = pd.read_csv("Base_de_datos.csv")
```

```python
datos
```

```python
SLV = datos[datos.country == "El Salvador"]
```

```python
SLV
```

```python
GTM = datos[datos.country == "Guatemala"]
```

```python
GTM
```

```python
HND = datos[datos.country == "Honduras"]
```

```python
HND
```

```python
PAN = datos[datos.country == "Panama"]
```

```python
PAN
```

```python
plt.plot(SLV.year, SLV.population/10**6)
plt.plot(GTM.year, GTM.population/10**6)
plt.plot(HND.year, HND.population/10**6)
plt.plot(PAN.year, PAN.population/10**6)
plt.legend(["El Salvador", "Guatemala", "Honduras", "Panama"])
plt.xlabel("Año")
plt.ylabel("Poblacion (millones)")
plt.show()
```
